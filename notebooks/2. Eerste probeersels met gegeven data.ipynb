```python
import random
k=random.randint(0,len(iterator))
m=iterator[k]
m
```

```python
from rdkit.Chem import AllChem

n = Chem.AddHs(iterator[k])
AllChem.Compute2DCoords(n)

n
```

```python
AllChem.EmbedMolecule(n)
```

```python
AllChem.UFFOptimizeMolecule(n)
n
```

```python
cids = AllChem.EmbedMultipleConfs(n, numConfs=10, pruneRmsThresh=1)
len(cids)

for cid in cids:
    ff = AllChem.MMFFGetMoleculeForceField(n, AllChem.MMFFGetMoleculeProperties(n), confId=cid)
    print(ff.CalcEnergy())
```
